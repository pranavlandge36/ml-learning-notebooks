```python
df=pd.read_csv('spam.tsv',sep='\t',names=['class','message'])
```

```python
import string
string.punctuation
```

```python
def remove_punc(text):
    text= ''.join([char for char in text if char not in string.punctuation])
    return text

df['clean']=df['message'].apply(lambda x : remove_punc(x))
```

```python
df
```

```python
x=df['clean']
y=df['class']
```

```python
y=y.astype(int)
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
x_cv=cv.fit_transform(x_train)
```

```python
x_test_cv=cv.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_cv,y_train)
```

```python
predict=model.predict(x_test_cv)
```

```python
from sklearn.metrics import classification_report, f1_score
```

```python
print(classification_report(y_test,predict))
print(f1_score(y_test,predict))
```

```python
msg=input('message')
msg_input=cv.transform([msg])
pre=model.predict(msg_input)
if pre==0:
    print('spam')
else :
    print('not spam')
```
