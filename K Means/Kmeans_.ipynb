```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
```

```python
df[['fy2017, Dealing with construction permits','fy2019, Dealing with construction permits','fy2018, Dealing with construction permits']]=scaling.fit_transform(df[['fy2017, Dealing with construction permits','fy2019, Dealing with construction permits','fy2018, Dealing with construction permits']])
```

```python
df
```

```python
x=df.drop('country',axis=1)
```

```python
x
```

```python
x.rename(columns={'fy2017, Dealing with construction permits':'fy2017',
                  'fy2019, Dealing with construction permits':'fy2019',
                  'fy2018, Dealing with construction permits':'fy2018'},inplace=True)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,5):
    kmeans=KMeans(n_clusters=i,random_state=29)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,5),wcss)
plt.xlabel('no. of clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans
model =KMeans(n_clusters=3,random_state=20)
model.fit(x)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
x
```

```python
x.rename(columns={'2017':'fy2017',
                  '2019':'fy2019',
                  '2018':'fy2018'},inplace=True)
```

```python

color_schema=np.array(['red','green','violet'])
plt.scatter(x.fy2017,x.fy2019,x.fy2018,color=color_schema[model.labels_])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    x['fy2017'],
    x['fy2019'],
    x['fy2018'],
    c=color_schema[model.labels_],
    s=30,
    depthshade=True
)

ax.set_xlabel('fy2017')
ax.set_ylabel('fy2019')
ax.set_zlabel('fy2018')
plt.title('3D scatter (2017, 2019, 2018) colored by cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,model.labels_)
```
